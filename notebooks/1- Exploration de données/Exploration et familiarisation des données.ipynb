# **Combine les données**

Relationship.JPG

```python
merge_data = pd.merge(bureau, customer, on = 'cust_id', how = 'inner')
final_data = pd.merge(merge_data, loan, on = 'cust_id', how = 'inner')

final_data.head()
```

```python
final_data.columns
```

```python
df = final_data.drop('loan_id', axis = 1)
df.head()
```

# **Explorer les colonnes catégoriques**

```python
cat_cols = ['gender', 'marital_status', 'employment_status', 'residence_type', 'loan_purpose', 'loan_type', 'default']

fig, axes = plt.subplots(3, 3, figsize=(18, 12))
axes = axes.flatten()

for i, column in enumerate(cat_cols):
    sns.countplot(data=df, x=column, ax=axes[i], palette='Set2')
    axes[i].set_title(f"Distribution de {column.capitalize()}")

# Remove empty subplots (if any)
for j in range(len(cat_cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

**Résumé des constats**

Les graphiques montrent la distribution de plusieurs variables catégorielles dans le jeu de données.

* **Gender:** La majorité des individus du jeu de données sont des hommes.
* **Marital Status:** (Statut marital) Une proportion importante des personnes sont mariées.
* **Employment Status:** (Statut d’emploi) La plupart des individus sont travailleurs indépendants.
* **Residence Type:** (Type de résidence) La majorité des personnes sont propriétaires de leur logement.
* **Loan Purpose:** (Motif du prêt) Les prêts immobiliers et personnels sont les plus fréquents.
* **Loan Type:** (Type de prêt) Les prêts garantis sont plus répandus que les prêts non garantis.
* **Default:** (Défaut de paiement) Une large majorité des prêts ne sont pas en défaut.

# **Explorez la distribution des colonnes numériques**

```python
# Liste des colonnes numériques
num_cols = df.select_dtypes(include=['number']).columns

# Création d’une grille de sous-graphiques 8x3
fig, axes = plt.subplots(8, 3, figsize=(18, 36))  # Ajuster figsize si nécessaire
axes = axes.flatten()  # Aplatir la grille pour un indexage plus simple

# Tracer chaque colonne numérique
for i, column in enumerate(num_cols):
    sns.boxplot(data=df, y=column, ax=axes[i], palette='Set2')
    axes[i].set_title(f"Boîte à moustaches de {column.capitalize()}")

# Supprimer les sous-graphiques vides (le cas échéant)
for j in range(len(num_cols), len(axes)):
    fig.delaxes(axes[j])

# Ajuster la mise en page
plt.tight_layout()
plt.show()
```

**Résumé des constats**

Les boxplots mettent en évidence la distribution des différentes variables numériques du jeu de données.

* **Number of Open Accounts:** (Nombre de comptes ouverts) La majorité des individus possèdent entre 1 et 10 comptes ouverts.
* **Number of Closed Accounts:** (Nombre de comptes clôturés) La plupart des individus ont entre 0 et 10 comptes clôturés.
* **Total Months:** (Total de mois) La distribution est asymétrique à droite, avec une majorité d’individus entre 0 et 200 mois.
* **Delinquent Months:** (Mois de défaillance) La majorité des individus présentent 0 mois de défaillance.
* **Total Age:** (Âge total) La distribution est asymétrique à droite, avec une majorité d’individus âgés entre 20 et 60 ans.
* **Income:** (Revenu) La distribution est asymétrique à droite, avec une majorité d’individus gagnant moins de 100 000.
* **Number of Dependents:** (Nombre de personnes à charge) La plupart des individus ont 0 ou 1 personne à charge.
